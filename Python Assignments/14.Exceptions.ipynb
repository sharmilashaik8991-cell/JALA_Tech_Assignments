```python
# 1. Write a program to generate Arithmetic Exception without exception handling
num1 = 10
num2 = 0
result = num1 / num2
print("Result:", result)
```

```python
# 2. Handle the Arithmetic exception using try-catch block
num1 = 10
num2 = 0  # This will cause division by zero

try:
    result = num1 / num2
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed!")

# Program continues
print("Program continues after exception handling.")
```

```python
# 3. Write a method which throws exception, Call that method in main class without try block
# Method that raises an exception
def divide_numbers(a, b):
    if b == 0:
        raise ZeroDivisionError("Cannot divide by zero!")  # Explicitly throw exception
    return a / b

# Main program
if __name__ == "__main__":
    result = divide_numbers(10, 0)  # Call method without try block
    print("Result:", result)  # This line will not execute
```

```python
# 4. Write a program with multiple catch blocks
try:
    num = int(input("Enter a number: "))
    result = 10 / num
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Division by zero!")
except ValueError:
    print("Error: Invalid input, please enter a number!")

print("Program continues after multiple catch blocks.")
```

```python
# 5. Write a program to throw exception with your own message
def check_age(age):
    if age < 18:
        raise Exception("Age must be 18 or older!")  # Custom message
    else:
        print("Age is valid.")

# Main
check_age(15)  # Will throw exception
```

```python
# 6. Write a program to create your own exception
# Custom exception class
class MyCustomError(Exception):
    pass

def divide(a, b):
    if b == 0:
        raise MyCustomError("Cannot divide by zero!")
    return a / b

# Main
divide(10, 0)
```

```python
# 7. Write a program with finally block
try:
    num = int(input("Enter a number: "))
    result = 10 / num
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Division by zero!")
finally:
    print("This will always execute, whether exception occurs or not.")
```
