```python
# 1. Different ways creating a string
str1 = 'Hello'                 # Using single quotes
str2 = "World"                 # Using double quotes
# Using triple quotes (for multi-line strings)
str3 = '''This is 
a multi-line                   
string'''
print(str1)  
print(str2) 
print(str3)  
```

```python
# 2. Concatenating two strings using + operator
str1 = "Hello"
str2 = "World"
result = str1 + " " + str2
print(result)  
```

```python
# 3. Finding the length of the string
A= "Hello World"
length = len(A)
print(length)  
```

```python
# 4. Extract a string using Substring
text = "Hello World"
substring = text[0:5]  # takes index from 0 to 4(5 does not included)
print(substring)  

substring2 = text[6:]  # takes index from index 6 to end(because at last we didn't represent any number so it moves upto last)
print(substring2) 
```

```python
# 5. Searching in strings using index()
Z = "Hello World"
position_1 = Z.index("W")
position_2= Z.index("Hello")       #even though we gave a complete word or complete sentence, when we use index to show it shows only one value index 
print(position_1)                  #because index has a capability to show only one index number
print(position_2) 
```

```python
# 6. Matching a String Against a Regular Expression With matches()
import re
text = "Hello123"
pattern = r'^[A-Z a-z 0-9]+$'  # Only letters and numbers
match = re.fullmatch(pattern, text)
print(match is not None)  
```

```python
# 7. Comparing strings
str1 = "apple"
str2 = "banana"
print(str1 == str2)  # checks equality but it doesn't have eqaulity
print(str1 != str2)  # checks inequality 

a=(1,2,3,4,5)
b=(1,2,3,4,5)
c=(1,2,5,4,3)
print(a==b)
print(a==c)
```

```python
# 8. startsWith(), endsWith() and compareTo()
text = "Hello World"
print(text.startswith("Hello")) 
print(text.endswith("World")) 
print(text.startswith("H"))
print(text.startswith("W"))

# Python doesn't have compareTo so we are using comparison operators
print("apple" < "banana")  # Comparing according to lexicographical order)
print("apple" > "banana") 
```

```python
# 9. Trimming strings with strip()
text = "   Hello World   "
trimmed = text.strip()                  #avoids spaces both front and back
print(trimmed)  
```

```python
# 10. Replacing characters in strings with replace()
text = "Hello World"
after_replace = text.replace("World", "Python")
print(after_replace)  
replace=text.replace("Hello","hai")
print(replace)
```

```python
# 11. Splitting strings with split()
text = "apple,banana,cherry"
print(text) 
fruits = text.split(",")
print(fruits)
```

```python
# 12. Converting integer objects to Strings
num = 123
num_str = str(num)
print(num)
print(num_str)       
print(type(num_str)) 
```

```python
# 13. Converting to uppercase and lowercase
text = "Hello World"
print(text.upper()) 
print(text.lower()) 
```
