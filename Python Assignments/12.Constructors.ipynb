```python
# 1. Write a class with a default constructor, one argument constructor and two argument constructors. 
#Instantiate the class to call all the constructors of that class from a mainclass
class Person:
    # Constructor with default arguments
    def __init__(self, name=None, age=None):
        if name is not None and age is not None:
            print(f"Two-argument constructor called: Name = {name}, Age = {age}")
        elif name is not None:
            print(f"One-argument constructor called: Name = {name}")
        else:
            print("Default constructor called")

# Main class / main section
if __name__ == "__main__":
    # Call default constructor
    p1 = Person()

    # Call one-argument constructor
    p2 = Person("Alice")

    # Call two-argument constructor
    p3 = Person("Bob", 25)
```

```python
# 2. Call the constructors(both default and argument constructors) of super class from a child class
# Superclass
class Person:
    # Constructor with default and argument options
    def __init__(self, name=None, age=None):
        if name is not None and age is not None:
            print(f"Person two-argument constructor called: Name = {name}, Age = {age}")
        elif name is not None:
            print(f"Person one-argument constructor called: Name = {name}")
        else:
            print("Person default constructor called")


# Child class
class Employee(Person):
    def __init__(self, name=None, age=None, emp_id=None):
        if name is None and age is None:
            # Call default constructor of superclass
            super().__init__()
        else:
            # Call argument constructor of superclass
            super().__init__(name, age)
        
        if emp_id is not None:
            print(f"Employee constructor called: Employee ID = {emp_id}")


# Main section
if __name__ == "__main__":
    print("Calling default constructor of super class from child class:")
    e1 = Employee()  # Calls default constructor of Person

    print("\nCalling argument constructor of super class from child class:")
    e2 = Employee("Alice", 30, emp_id=101)  # Calls argument constructor of Person
```

```python
# 3. Apply private, public, protected and default access modifiers to the constructor
class MyClass:
    # Public constructor
    def __init__(self):
        print("Public constructor called")

    # Protected constructor (convention)
    def _protected_init(self):
        print("Protected constructor called")

    # Private constructor (name mangling)
    def __private_init(self):
        print("Private constructor called")

# Subclass to test access
class ChildClass(MyClass):
    def access_parent(self):
        print("Accessing public constructor from subclass:")
        self.__init__()  # Public → works fine

        print("Accessing protected constructor from subclass:")
        self._protected_init()  # Protected → works by convention

        # Private constructor cannot be accessed directly
        print("Trying to access private constructor from subclass:")
        try:
            self.__private_init()
        except AttributeError as e:
            print("Error:", e)


# Main
if __name__ == "__main__":
    obj = ChildClass()
    
    print("\nAccessing constructors through subclass object:")
    obj.access_parent()

    print("\nTrying to access private constructor from main:")
    try:
        obj.__private_init()
    except AttributeError as e:
        print("Error:", e)
```

```python
# 4. Write a program which illustrates the concept of attributes of a constructor
class Person:
    # Constructor with attributes
    def __init__(self, name, age, country="Unknown"):
        self.name = name      # Attribute: name
        self.age = age        # Attribute: age
        self.country = country  # Attribute with default value

    # Method to display attributes
    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}, Country: {self.country}")


# Main section
if __name__ == "__main__":
    # Create objects with different attributes
    p1 = Person("Alice", 25, "USA")     # Providing all attributes
    p2 = Person("Bob", 30)              # Country defaults to "Unknown"

    print("Object 1 attributes:")
    p1.display_info()

    print("\nObject 2 attributes:")
    p2.display_info()
```
