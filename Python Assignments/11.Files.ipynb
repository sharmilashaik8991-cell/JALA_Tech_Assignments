```python
# 1. Write a program to read text file
# Open the file in read mode
file = open("example.txt", "r")  # "r" means read mode

# Read the entire content of the file
content = file.read()

# Print the content
print("File Content:")
print(content)

# Close the file
file.close()
```

```python
# 2. Write a program to write text to .txt file using InputStream
# Open file in write mode ("w") - this will create the file if it does not exist
with open("output.txt", "w") as file:
    # Take input from user
    text = input("Enter text to write to file: ")

    # Write the input text to the file
    file.write(text)

print("Text has been written to output.txt successfully.")
```

```python
# 3. Write a program to read a file stream
# Open file in read mode
with open("example.txt", "r") as file:
    # Read the file content
    content = file.read()
    print("File content:")
    print(content)
```

```python
# 4. Write a program to read a file stream supports random access
# Open file in read mode
with open("example.txt", "r") as file:
    print("Current file position:", file.tell())  # Starting at 0
    
    # Read first 20 characters
    first_part = file.read(20)
    print("First 20 characters:", first_part)
    
    # Random access: jump to 10th byte
    file.seek(10)
    print("Position after seek(10):", file.tell())
    
    # Read next 15 characters from position 10
    next_part = file.read(15)
    print("Next 15 characters after seek:", next_part)
```

```python
# 5. Write a program to read a file a just to a particular index using seek()
with open("example.txt", "r") as file:
    # Move to position 0
    file.seek(0)
    
    # Read up to 30th character
    data = file.read(30)
    print("First 30 characters of the file:")
    print(data)
```

```python
# 6. Write a program to check whether a file is having read access and write access permission
import os

filename = "example.txt"

# Check read permission
if os.access(filename, os.R_OK):
    print(f"{filename} is readable")
else:
    print(f"{filename} is not readable")

# Check write permission
if os.access(filename, os.W_OK):
    print(f"{filename} is writable")
else:
    print(f"{filename} is not writable")
```
