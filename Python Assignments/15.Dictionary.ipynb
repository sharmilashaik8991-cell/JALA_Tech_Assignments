```python
# 1. Create a Dictionary with at least 5 key value pairs of the Student ID and Name
Student_details={"ID":206,"Name":"Sharmila","Course":"B.Tech","mail":"sharmilsahaik8991@gmail.com","contact":6301237819}
print(Student_details)
print(Student_details.keys())
print(Student_details.values())
```

```python
# 1.1. Adding the values in dictionary
Student_details["college"]="Engineering College"
print(Student_details)
```

```python
# 1.2. Updating the values in dictionary
Student_details["Course"]="M.Tech"
print(Student_details)
```

```python
# 1.3. Accessing the value in dictionary
print(Student_details["ID"]) 
print(Student_details["Name"]) 
print(Student_details["Course"]) 
print(Student_details["mail"]) 
print(Student_details["contact"])
```

```python
# 1.4. Create a nested loop dictionary
# Outer dictionary: each student ID maps to another dictionary
Students = {
    101: {
        "Name": "Sharmila",
        "Course": "B.Tech",
        "Mail": "sharmilsahaik8991@gmail.com",
        "Contact": 6301237819
    },
    102: {
        "Name": "Rohit",
        "Course": "B.Sc",
        "Mail": "rohit123@gmail.com",
        "Contact": 9876543210
    },
    103: {
        "Name": "Anjali",
        "Course": "M.Tech",
        "Mail": "anjali@gmail.com",
        "Contact": 9123456780
    }
}
print(Students)
```

```python
# 1.5. Access the values of nested loop dictionary
print(Students[101]["Name"]) 
print(Students[101]["Course"])  
print(Students[101]["Mail"])    
print(Students[101]["Contact"]) 
print(Students[102]["Name"]) 
print(Students[102]["Course"]) 
print(Students[102]["Mail"]) 
print(Students[102]["Contact"]) 
print(Students[103]["Name"]) 
print(Students[103]["Course"]) 
print(Students[103]["Mail"]) 
print(Students[103]["Contact"]) 
```

```python
# 1.6. Print the keys present in a particular dictionary
Student_details={"ID":206,"Name":"Sharmila","Course":"B.Tech","mail":"sharmilsahaik8991@gmail.com","contact":6301237819}
print(Student_details.keys())
```

```python
# 1.7. Delete a value from a dictionary
Student_details.pop("contact")
print(Student_details)
```
