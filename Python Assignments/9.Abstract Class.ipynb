```python
# 1. Create an abstract class with abstract and non-abstract methods.
#In Python, we use the abc module to define abstract classes and abstract methods.
from abc import ABC, abstractmethod

# Abstract class
class Vehicle(ABC):

    # Abstract method: must be implemented by any subclass
    @abstractmethod
    def start_engine(self):
        pass

    # Non-abstract method: can be used directly by subclass
    def fuel_type(self):
        print("This vehicle runs on petrol, diesel, or electricity.")
# Subclass that implements the abstract method
class Car(Vehicle):
    def start_engine(self):
        print("Car engine started!")  # Implementation of abstract method

# Create an instance of the subclass
my_car = Car()

# Call the abstract method (implemented in subclass)
my_car.start_engine()

# Call the non-abstract method (inherited from Vehicle)
my_car.fuel_type()
```

```python
#2. Create a sub class for an abstract class. Create an object in the child class for the abstract class and access the non-abstract methods
from abc import ABC, abstractmethod
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

    def fuel_type(self):
        print("This vehicle runs on petrol, diesel, or electricity.")

# Subclass
class Car(Vehicle):
    def start_engine(self):
        print("Car engine started!")  # Implementing the abstract method

    # Method to access non-abstract method from parent
    def show_fuel(self):
        print("Accessing non-abstract method from Vehicle:")
        super().fuel_type()  # Calls fuel_type() from Vehicle


# Create an instance of the child class
my_car = Car()

# Call child method that accesses non-abstract parent method
my_car.show_fuel()
```

```python
#3. Create an instance for the child class in child class and call abstract methods
from abc import ABC, abstractmethod

# Abstract class
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass  # Abstract method must be implemented

# Child class
class Car(Vehicle):
    def start_engine(self):
        print("Car engine started!")  # Implementing abstract method

# Create instance of child class
my_car = Car()

# Call the abstract method (implemented in child class)
print("Step 3: Calling abstract method using child instance:")
my_car.start_engine()
```

```python
#4. Create an instance for the child class in child class and call non-abstract method
from abc import ABC, abstractmethod

# Abstract class
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass  # Abstract method must be implemented

    def fuel_type(self):
        print("This vehicle runs on petrol, diesel, or electricity.")  # Non-abstract method

# Child class
class Car(Vehicle):
    def start_engine(self):
        print("Car engine started!")  # Implementing abstract method

# Create instance of child class
my_car = Car()

# Call the non-abstract method (inherited from abstract class)
print("Step 4: Calling non-abstract method using child instance:")
my_car.fuel_type()
```
