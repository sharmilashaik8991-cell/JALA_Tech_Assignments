```python
# 1. Define a static variable and access that through a class
class MyClass:
    static_var = 10         #normally this is called as local variable but by using class name.variable name we can call anywhere we want
print("Access through class:", MyClass.static_var)
```

```python
# 2. Define a static variable and access that through a instance
class MyClass:
    static_var = 10
obj = MyClass()               # Creating an instance
print("Access through instance:", obj.static_var)      #accesses the class variable
```

```python
# 3. Define a static variable and change within the instance
class MyClass:
    static_var = 10
obj = MyClass()                   # Create an instance
obj.static_var = 20               # Changing the static variable using the instance
print("Instance after change:", obj.static_var)             # Access through instance
print("Class after instance change:", MyClass.static_var)   # Access through class
```

```python
# 4. Define a static variable and change within the class
class MyClass:
    static_var = 10
MyClass.static_var = 30           # Changing static variable through the class

# Creating instances
obj1 = MyClass()
obj2 = MyClass()

# Accessing through instances
print("Instance 1:", obj1.static_var)
print("Instance 2:", obj2.static_var)

# Accessing through class
print("Class:", MyClass.static_var)
```
