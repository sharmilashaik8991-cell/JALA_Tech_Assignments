```python
# 10. Write a function to find the duplicate values of an array
# Define the array
a = [1, 2, 3, 2, 4, 5, 3, 6, 4]
def find_duplicates(arr):                             # Function to find duplicate values
    duplicates = []   # To store duplicates
    seen = set()      # To keep track of numbers we have already checked

    for num in arr:
        if num in seen and num not in duplicates:     # If num is already seen, it is a duplicate
            duplicates.append(num)
        seen.add(num)                                 # Add number to seen set
    return duplicates
print("Array:", a)
print("Duplicate values:", find_duplicates(a))
```

```python
# 11. Write a program to find the common values between two arrays
arr1 = [1, 2, 3, 4, 5]
arr2 = [3, 4, 5, 6, 7]
def common_values(a1, a2):           # Function to find common values
    return list(set(a1) & set(a2))   # Convert both lists to sets and find intersection
print("Array 1:", arr1)
print("Array 2:", arr2)
print("Common values:", common_values(arr1, arr2))
```

```python
# 13. Write a method to find the second largest number in an array
arr = [5, 2, 8, 1, 8, 4]
def second_largest(a):
    unique_values = list(set(a))   # Remove duplicates
    unique_values.sort()           # Sort in ascending order
    if len(unique_values) >= 2:
        return unique_values[-2]   # Second last element is second largest
    else:
        return None                # If array has less than 2 unique numbers
print("Array:", arr)
print("Second largest number:", second_largest(arr))
```

```python
# 15. Write a method to find number of even number and odd numbers in an array
arr = [1, 2, 3, 4, 5, 6, 7]
def count_even_odd(a):
    even = 0
    odd = 0
    for num in a:
        if num % 2 == 0:
            even += 1
        else:
            odd += 1
    return even, odd

even_count, odd_count = count_even_odd(arr)
print("Array:", arr)
print("Even numbers:", even_count)
print("Odd numbers:", odd_count)
```
