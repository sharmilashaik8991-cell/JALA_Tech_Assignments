```python
# 1. Write a function for arithmetic operators(+,-,*,/)
a=int(input("enter a:"))
b=int(input("enter b:"))
print("addition:",a+b)
print("subtraction:",a-b)
print("multiplication:",a*b)
print("division:",a/b)
```

```python
# 2. Write a method for increment and decrement operators(++, --)
def increment(num):
    num+=1
    return num
def decrement(num):
    num-=1
    return num
x=10
print(x)
print(increment(x))
print(decrement(x))

#or
a=5
print(a)
a+=1
print(a)
a-=1
print(a)
```

```python
# 3. Write a program to find the two numbers equal or not.
a=int(input("enter a:"))
b=int(input("enter b:"))
if (a==b):
    print("two numbers are equal")
else:
    print("two numbers are not equal")
```

```python
# 4. Program for relational operators (<,<==, >, >==)
a=int(input("enter a:"))
b=int(input("enter b:"))
print("a<b:",a<b)
print("a<b:",a>b)
print("a<b:",a<=b)
print("a<b:",a<=b)
```

```python
# 5. Print the smaller and larger number
a=int(input("enter a:"))
b=int(input("enter b"))
if a>b:
    print("a is larger:",a)
    print("b is smaller:",b)
elif a<b:
    print("a is smaller:",a)
    print("b is larger:",b)
else:
    print("both are equal")
```
