```python
# 8. Write a program to find Armstrong number or not
a=int(input("enter number:"))
b=a                               #it stores the original number a into b bcz we need the original number later to check if it’s Armstrong.
number_of_digits=len(str(a))      #Converts a to a string to count its digits.Without this, we wouldn’t know what power to raise each digit to for numbers with more than 3 digits.
sum_of_powers=0                   #Initializes a variable to store the sum of each digit raised to the power of number_of_digits.
while a>0:                        #loop iteratea until it became false
    digit=a% 10                   #% gives the remainder when dividing by 10 → gives the last digit of a.
    sum_of_powers+=digit**number_of_digits
    a//=10
if sum_of_powers==b:
    print(f"{b} is an Armstrong number")
else:
    print("%d is not an Armstrong number.\n",b)
    
    
#taken number and making osmething number into cube should be same then it is called armstrong number
#153=1cube+5cube+3cube=1+125+27=153
```

```python
# 9. Write a program to find the prime or not.
a=int(input("enter number"))
if a<=1:                         #it is taken bcz to eliminate 0 and 1 bcs they are not prime numbers
    print(f"{a} is not a prime number.")
else:
    is_prime = True              # Assume the number is prime
    for i in range(2, int(a ** 0.5) + 1):  # Check divisibility up to sqrt(a)
        if a % i == 0:
            is_prime = False
            break

    if is_prime:
        print(f"{a} is a prime number.")
    else:
        print(f"{a} is not a prime number.") 
```

```python
# 10. Write a program to palindrome or not
num = int(input("Enter a number: "))
b = num                     # Stores original number like question paper(unchanged)
reversed_num = 0            #initializing a variable to store reversed number

while num > 0:              #Loop used to reverse the number.
    digit = num % 10        # Get last digit
    reversed_num = reversed_num * 10 + digit              #Add the digit to the reversed number in the correct position.
    num //= 10                # Remove last digit

# Check if original and reversed numbers are same
if b == reversed_num:
    print(f"{b} is a palindrome.")
else:
    print(f"{b} is not a palindrome.")
```

```python
num = input("Enter a number: ")             #here num is treating as a string not an integer
if num == num[::-1]:                        
#here we used string slicing(to make reverse)  slicing= string[start : stop : step]. start and stop are  not given and step is given to move backwards
#and it compares original string with reversed string
    print("It is a palindrome.")           
else:
    print("It is not a palindrome.")
```
