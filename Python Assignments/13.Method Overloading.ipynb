```python
# 1. Write two methods with the same name but different number of parameters of same type and call the methods
class Calculator:
    def add(self, *args):
        if len(args) == 1:
            print(f"One parameter addition: {args[0]} + 10 = {args[0] + 10}")
        elif len(args) == 2:
            print(f"Two parameter addition: {args[0]} + {args[1]} = {args[0] + args[1]}")
        else:
            print("Invalid number of arguments")


# Main section
if __name__ == "__main__":
    calc = Calculator()

    # Call method with one parameter
    calc.add(5)

    # Call method with two parameters
    calc.add(5, 15)
```

```python
# 2. Write two methods with the same name but different number of parameters of different data type and call the methods
#In Python, you cannot create two methods with the same name even if their parameter types are different—the second method will overwrite the first one.
class Calculator:
    def multiply(self, *args):
        # One argument → check type
        if len(args) == 1:
            if isinstance(args[0], int):
                print(f"Integer multiplication: {args[0]} * 10 = {args[0] * 10}")
            elif isinstance(args[0], float):
                print(f"Float multiplication: {args[0]} * 1.5 = {args[0] * 1.5}")
            else:
                print("Unsupported type for single argument")

        # Two arguments → check type
        elif len(args) == 2:
            if all(isinstance(x, int) for x in args):
                print(f"Two-integer multiplication: {args[0]} * {args[1]} = {args[0] * args[1]}")
            elif all(isinstance(x, float) for x in args):
                print(f"Two-float multiplication: {args[0]} * {args[1]} = {args[0] * args[1]}")
            else:
                print("Mixed or unsupported types for two arguments")
        else:
            print("Invalid number of arguments")


# Main
if __name__ == "__main__":
    calc = Calculator()

    # Call method with single integer
    calc.multiply(5)

    # Call method with single float
    calc.multiply(2.5)

    # Call method with two integers
    calc.multiply(3, 4)

    # Call method with two floats
    calc.multiply(1.2, 3.4)
```

```python
# 3. Write two methods with the same name and same number of parameters of same type
class Calculator:
    # First method
    def add(self, a, b):
        print(f"First add method: {a} + {b} = {a + b}")

    # Second method with same name and same parameters
    def add(self, a, b):
        print(f"Second add method: {a} + {b} = {a + b + 10}")


# Main
if __name__ == "__main__":
    calc = Calculator()

    # Call add method
    calc.add(5, 10)
```
